```python
#
# required libraries
#
import pandas as pd
import datetime
import matplotlib.pyplot as plt
#
```

```python
#
# a little Jupyter trick here, so that we can 
# output multiple objects per cell without having to 
# use print() or display()
#
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
#
# load data from a .csv file to a DataFrame
#
fname = 'Datasets/test_data_frame.csv'
my_data = pd.read_csv(fname)
my_data.head()
my_data.tail()
```

```python
my_data.dtypes
```

```python
#
# convert the timestamps from object to a pandas time type
# in this case, we'll use timedelta
#
my_data['time'] = pd.to_timedelta(my_data['time'])
my_data.dtypes
```

```python
#
# now that the time columnn is a pandas time type
# let's replace the index with it
#
my_data.set_index('time', drop = True, inplace = True)
my_data.head()
my_data.tail()
```

```python
#
# now that we have a time-based index, we can apply some 
# powerful Pandas methods
# .resample() allows us to change the frequency of the
# observations, and applying .interpolate() fills the 
# new time slots with interpolated values
#
my_data = my_data.resample('5min').interpolate()
my_data.head()
my_data.tail()
```

```python
#
# we can summarize the data by the months by taking the column means
#
summary_data = my_data.mean(axis = 0)
summary_data
summary_data.shape
```

```python
#
# let's plot the data
# note that since the index labels are strings
# we rotate them or they might overlap on the plot
#
fig, ax = plt.subplots()
ax.plot(summary_data.index, summary_data)
ax.set_xticklabels(labels = summary_data.index, rotation = 90)
plt.show()
```

```python
# 
# let's simulate the case where we were missing
# July and August in the data
#
points = [0, 1, 2, 3, 4, 5, 8, 9, 10, 11]
subset_summary = summary_data[points]
subset_summary
fig, ax = plt.subplots()
ax.plot(subset_summary.index, subset_summary)
ax.set_xticklabels(labels = summary_data.index, rotation = 90)
plt.show()
```

We have a problem here--since the index is actually just character strings, the plot data are incorrectly labeled on the x-axis, and the spacing is wrong. We can correct this by using a Pandas datetime index.  Note that when we plot, we don't have to rotate the labels (if we don't want to) as matplotlib now interprets the index as dates, and takes care of a reasonable spacing for us.

```python
summary_data.index = pd.to_datetime(summary_data.index)
summary_data
fig, ax = plt.subplots()
ax.plot(summary_data.index, summary_data)
plt.show()
```

```python
#
# let's try the missing months simulation again
#
points = [0, 1, 2, 3, 4, 5, 8, 9, 10, 11]
subset_summary = summary_data[points]
fig, ax = plt.subplots()
ax.plot(subset_summary.index, subset_summary)
plt.show()
```

Voila!  Now the missing data are represented correctly (although this would be better represented as a scatter plot now!)
